# Código para Descargar Información Financiera de Yahoo

## 1. Configuración inicial

Se necesita importar la libreria `yfinance` la cual permite obtener financiera de Yahoo Finance.

```python
import yfinance as yf
import os
```

## 2. Obtener datos para la acción de apple.

Utilizando la función `yf.Ticker()` se selecciona el ticker de apple, a partir de la función `history()` obtenemos la información histórica de ese ticker. El parámetro `period = max`, trae toda la información histórica disponible.

```python
ticker = yf.Ticker("AAPL")
df = ticker.history(period="max")

df = df.reset_index()
df["Date"] = df["Date"].dt.tz_localize(None)
```

## 3. Guardamos la información descargada en la carpeta `data\raw`

```python
carpeta = os.getcwd().split("\\")[-1].upper()
while carpeta != "202610_DERIVADOS_TALLER1":
    os.chdir("..")
    carpeta = os.getcwd().split("\\")[-1].upper()
```

```python
output_dir = os.path.join(os.getcwd(), "data", "raw")
os.makedirs(output_dir, exist_ok=True)
output_path = os.path.join(output_dir, "AAPL_price.xlsx")
df.to_excel(output_path, index=False)
```
